```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from numpy.random import uniform
from sklearn.datasets import make_blobs
import seaborn as sns
import random


def most_common(lst):
    """
    Return the most frequently occurring element in a list.
    """
    return max(set(lst), key=lst.count)


def euclidean(point, data):
    """
    Return Euclidean distances between a point & a dataset.
    """
    return np.sqrt(np.sum((point - data) ** 2, axis=1))


class KMeans:
    def __init__(self, n_clusters=8, max_iter=300):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.centroids = None

    def fit(self, X_train):
        """
        Fit the KMeans model to the training data using k-means++ initialization.
        """
        # Initialize centroids with k-means++
        self.centroids = [random.choice(X_train)]
        for _ in range(self.n_clusters - 1):
            dists = np.sum([euclidean(c, X_train) for c in self.centroids], axis=0)
            dists /= np.sum(dists)  # normalize
            new_centroid_idx = np.random.choice(range(len(X_train)), p=dists)
            self.centroids.append(X_train[new_centroid_idx])

        iteration = 0
        prev_centroids = [np.zeros_like(c) for c in self.centroids]

        # Iterate until convergence or max_iter
        while iteration < self.max_iter and not np.allclose(self.centroids, prev_centroids):
            # Assign points to nearest centroid
            clusters = [[] for _ in range(self.n_clusters)]
            for x in X_train:
                dists = euclidean(x, self.centroids)
                centroid_idx = np.argmin(dists)
                clusters[centroid_idx].append(x)

            # Update centroids
            prev_centroids = self.centroids
            self.centroids = [
                np.mean(cluster, axis=0) if len(cluster) > 0 else prev_centroids[i]
                for i, cluster in enumerate(clusters)
            ]
            iteration += 1

    def evaluate(self, X):
        """
        Assign points in X to nearest centroids.
        Returns: (list of centroid coordinates, list of centroid indices).
        """
        centroids = []
        centroid_idxs = []
        for x in X:
            dists = euclidean(x, self.centroids)
            centroid_idx = np.argmin(dists)
            centroids.append(self.centroids[centroid_idx])
            centroid_idxs.append(centroid_idx)
        return centroids, centroid_idxs


# ==== Demo ====

# Create a dataset of 2D distributions
centers = 5
X_train, true_labels = make_blobs(n_samples=100, centers=centers,random_state=42)
X_train = StandardScaler().fit_transform(X_train)

# Fit KMeans
kmeans = KMeans(n_clusters=centers)
kmeans.fit(X_train)

# Evaluate classification
class_centers, classification = kmeans.evaluate(X_train)

# Plot results
sns.scatterplot(
    x=[x[0] for x in X_train],
    y=[x[1] for x in X_train],
    hue=true_labels,
    style=classification,
    palette="deep",
    legend=None
)
plt.plot(
    [c[0] for c in kmeans.centroids],
    [c[1] for c in kmeans.centroids],
    '+',
    markersize=12,
    markeredgewidth=3,
    color="black"
)
plt.title("Custom K-Means Implementation")
plt.show()







#Stability Analysis:

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import adjusted_rand_score, silhouette_score
import numpy as np

def clustering_stability(X, n_clusters=5, n_runs=10, perturb=False, noise_level=0.01):
    """
    Perform stability analysis for custom KMeans.
    """
    clusterings = []
    silhouette_scores = []
    
    for run in range(n_runs):
        # Perturb data if required
        if perturb:
            X_run = X + np.random.normal(0, noise_level, X.shape)
        else:
            X_run = X.copy()
        
        # Fit custom KMeans
        kmeans = KMeans(n_clusters=n_clusters)
        kmeans.fit(X_run)
        _, labels = kmeans.evaluate(X_run)
        
        clusterings.append(labels)
        
        # Compute silhouette score (internal quality metric)
        sil = silhouette_score(X_run, labels)
        silhouette_scores.append(sil)
    
    # Pairwise Adjusted Rand Index (stability across runs)
    stability_scores = []
    for i in range(len(clusterings)):
        for j in range(i+1, len(clusterings)):
            ari = adjusted_rand_score(clusterings[i], clusterings[j])
            stability_scores.append(ari)
    
    return stability_scores, silhouette_scores


# ==== Run stability tests ====
stability_scores, silhouette_scores = clustering_stability(X_train, n_clusters=centers, n_runs=10)

# ==== Visualization ====
plt.figure(figsize=(12, 5))

# ARI boxplot
plt.subplot(1, 2, 1)
sns.boxplot(y=stability_scores, color="skyblue")
plt.title("Stability Across Runs (ARI)")
plt.ylabel("Adjusted Rand Index (0  1)")
plt.axhline(np.mean(stability_scores), color="red", linestyle="--", label="Mean ARI")
plt.legend()

# Silhouette boxplot
plt.subplot(1, 2, 2)
sns.boxplot(y=silhouette_scores, color="lightgreen")
plt.title("Clustering Quality (Silhouette Scores)")
plt.ylabel("Silhouette Score (-1  1)")
plt.axhline(np.mean(silhouette_scores), color="red", linestyle="--", label="Mean Silhouette")
plt.legend()

plt.suptitle("KMeans Stability Analysis", fontsize=14)
plt.tight_layout()
plt.show()
```
